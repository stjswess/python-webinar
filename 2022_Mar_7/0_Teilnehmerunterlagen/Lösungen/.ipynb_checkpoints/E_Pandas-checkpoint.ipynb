**(A-1)** Lade die Datei *"Eisverkauf_Sommer_2019.csv"* in ein DataFrame. Du findest diese im Ordner der Aufgaben. *Achte auf das richtige Trennzeichen!*

* Wandele die Datumsspalte `Tag` in ein Datum (`datetime64`) um
* Wie viele fehlende Werte gibt es? Entferne diese!
* Wie viele Kugeln Eis wurden zwischen den `01.08 - 30.09.`verkauft?
* An welchen Tag wurde am wenigsten verkauft?


*Zusatz* Berechne die Korrelation zwischen dem Eisverkauf und der Temperatur

```python
# Laden der Daten
df = pd.read_csv("Eisverkauf_Sommer_2019.csv", sep=";")

# Umwandeln der Spalte
df["Tag"] = df["Tag"].astype("datetime64")

# Ausgabe fehlende Werte
print(df.isnull().sum())

# Entfernen fehlender Werte
df.dropna(axis=0, inplace=True)
print(df.isnull().sum())

# Selektieren des Zeitraums
df_zeitraum = df[(df["Tag"] >= pd.to_datetime("01.08.2019")) & (df["Tag"] <= pd.to_datetime("30.09.2019"))]
eis_kugeln = df_zeitraum["Kugeln_Eis"].sum()
print(f'Es wurde {eis_kugeln} verkauft!')

# Print Ausgabe des Tages mit dem geringsten Verkauf
df.loc[df["Kugeln_Eis"] == min(df["Kugeln_Eis"]),"Tag"]
```

**(A-2)**  Verkaufszahlen Produkte 
Und hier kommt dein erstes Datenpaket! Löse folgende Aufgaben, um es zu analysieren.
1. Lade die Daten `Verkaufszahlen.xlsx` in einem DataFrame.
1. Erstelle eine Statistik über die Werte
2. Ersetze fehlende Werte mit dem Spaltenmittelwert 
3. Sortiere Ausreißer der Spalte `Umsatz Kopfhörer`aus (Bsp. mit Quantilen)
4. Entwerfe eine Spalte "Gesamtumsatz" und addiere den Umsatz aller Produkte
5. Oh nein! Das Produkt "Apfel" hat eine Mehrwertsteuer von 7% statt 19%, korrigiere die Werte der Spalte 
6. Gebe die kumulierte Jahressumme von 2018 aus
7. Welcher Umsatz ist in den einzelnen Monaten zu erwarten? Welcher Umsatz wurde über alle Jahre hinweg durchschnittlich erzielt?(Benutze hierfür [`df.groupby()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) auf einer Spalte nur mit dem Monatswert)

|         | Umsatz Apfel |  Umsatz Hose |  Umsatz Kopfhörer |
|---------|-------|------|-----------|
| Januar  | ...   | ...  | ...       |
| Februar | ...   | ...  | ...       |
| März    | ...   | ...  | ...       |
| April   | ...   | ...  | ...       |

```python
# 1. Laden der Daten 
df = pd.read_excel("Verkaufszahlen.xlsx")

# 2. Erstellen der Statistik
df.describe()

# 3. Ersetzen fehlender Werte
df.fillna(value={"Umsatz Apfel": df["Umsatz Apfel"].mean(),
                 "Umsatz Hose": df["Umsatz Hose"].mean(),
                 "Umsatz Kopfhörer": df["Umsatz Kopfhörer"].mean()}, inplace=True)

# 4. Aussortieren der Ausreißer
unten = df["Umsatz Kopfhörer"].quantile(0.01)
oben = df["Umsatz Kopfhörer"].quantile(0.99)

df = df[(df["Umsatz Kopfhörer"] < oben) & (df["Umsatz Kopfhörer"] > unten)]
df.describe()

# 5. Berechnen des Gesamtumsatzes
df["Gesamtumsatz"] = df["Umsatz Apfel"] + df["Umsatz Hose"] + df["Umsatz Kopfhörer"]

# 6. Ändern der Mehrwertsteuer für Apfel von 19% auf 7%
df["Umsatz Apfel"] = df["Umsatz Apfel"] / 0.19 * 0.07

# 7. Kummulierte Jahressumme 2018
df.loc[df["Datum"].dt.year == 2018,"Gesamtumsatz"].cumsum()

# 8. Monatssumme
df["Monat"] = df["Datum"].dt.month
df[["Monat","Umsatz Apfel","Umsatz Hose","Umsatz Kopfhörer"]].groupby("Monat").mean()
```

Hier haben wir [diesen](https://www.kaggle.com/residentmario/ramen-ratings) entdeckt: einen 2580-teiligen Bewertungsatz von verschiedenen Ramen-Nudeln. In der Analyse haben sich einige Fehler eingeschlichen. Finde diese! *Achte auf die Kommentare, was das Ziel des Schrittes war!* (5 Fehler)

```python
# Einlesen und Info
#df= pd.read_csv("Ramen-Rating.csv",sep="\t")
df= pd.read_csv("Ramen-Rating.csv",sep=",")
print(df.info())

# Auffüllen von fehlenden Werten im DataFrame
#df.fillna(value=0)
df.fillna(value=0, inplace=True)

# Umwandeln der Stars "Unrated" in 0
df.loc[df["Stars"] == "Unrated","Stars"] = 0
#df["Stars"].astype('float64')
df["Stars"] = df["Stars"].astype('float64')

# Ausgeben der Top 5 der häufigsten Brands
werte = df["Brand"].value_counts()
print(werte.head(5))

# Ausgeben der am besten bewerteten Marken
#gruppiert_ratings = df["Stars"].groupby("Brand").sum()
gruppierte_ratings  = df[["Stars","Brand"]].groupby("Brand").mean()
#ratings_brand_top_10  = gruppierte_ratings.sort_values(by="Stars").head(10)
ratings_brand_top_10  = gruppierte_ratings.sort_values(by="Stars",ascending=False).head(10)
ratings_brand_top_10 

# Hinzufügen der Brandratings
df["Brand_Rating"] = df["Brand"].apply(lambda brand: gruppierte_ratings.loc[brand])
df
```
